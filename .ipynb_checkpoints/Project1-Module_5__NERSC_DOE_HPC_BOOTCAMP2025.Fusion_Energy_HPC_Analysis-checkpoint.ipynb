---

### üßë‚Äçüî¨ Why These Systems Matter for Fusion

1. **Nonlinear, Multiscale Physics**  
   Fusion simulations span **10+ orders of magnitude** in time and space. Capturing phenomena like plasma microturbulence, magnetic reconnection, and wall-material interactions requires exascale resources and advanced parallelism.

2. **Workload Diversity**  
   - **GPU-intensive ML** (e.g., plasma disruption predictors) excels on Perlmutter and Aurora.  
   - **PDE-heavy solvers** (e.g., gyrokinetic simulations) benefit from Frontier‚Äôs bandwidth and latency characteristics.  
   - **Multi-physics pipelines**, integrating simulation + diagnostics data, leverage El Capitan‚Äôs heterogeneous compute nodes.

3. **Energy Efficiency and Scaling**  
   DOE facilities increasingly optimize **science per watt**. Green500 metrics now influence allocation policies, especially for energy-intensive workflows like long-running plasma stability simulations.

4. **Real-time Experiment Integration**  
   Modern HPC workflows feed data from live fusion experiments into predictive models for **near-real-time control**‚Äîcritical for both tokamak operations and inertial confinement experiments.

## üá∫üá∏ National Strategy: The DOE Fusion Energy Vision

The **U.S. Department of Energy (DOE)** is pursuing one of the most ambitious scientific and engineering challenges of our time‚Äîbringing **commercial fusion energy** to reality. Guided by the **Fusion Energy Strategy 2024** and the **Fusion Energy Sciences (FES) Building Bridges Vision**, the DOE aims to transition fusion from experimental science to a deployable, carbon-free energy source within the next decade.

Fusion is no longer just a ‚Äúfuture hope‚Äù ‚Äî it‚Äôs on a strategic fast track. HPC systems, advanced diagnostics, and AI-driven modeling are critical enablers of this transition, and the DOE is placing high-performance computing at the heart of its strategy.

### üöÄ Key Strategic Pillars

- **Commercial Deployment by the 2030s**  
  Public-private partnerships will accelerate the development of pilot plants, creating pathways for industrial-scale deployment of fusion systems. This includes supporting companies designing compact, cost-efficient reactors, and aligning national laboratories with industry R&D timelines.

- **Closing Critical Technology Gaps**  
  Targeted research investments are focusing on:
  - **Materials under extreme conditions** (neutron irradiation, plasma-facing components)  
  - **Advanced magnetic confinement** for higher efficiency and stability  
  - **Tritium fuel cycle handling** for sustainable operation  
  - **Superconducting magnet innovation** for smaller, cheaper, and more efficient designs  

- **Integrated Research Infrastructure (IRI)**  
  Fusion research now operates in a **cyber-physical ecosystem**:
  - Experimental facilities (tokamaks, stellarators, laser-driven systems)  
  - HPC centers like **NERSC**, **OLCF**, and **ALCF** for real-time simulation and analysis  
  - **ESnet** for high-bandwidth, low-latency data transfer between experiment and computation  

- **Workforce Development & Equity Goals**  
  Building a next-generation fusion workforce that reflects the diversity of the nation is a priority. This includes DOE-supported fellowships, internships, and programs that bridge academia, industry, and national labs. Outreach and education initiatives aim to broaden participation in fusion research.

```python
## üìä Setting Up Fusion HPC Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Synthetic dataset based on realistic HPC workloads in 2025
df = pd.DataFrame({
    'JobID': range(1, 21),
    'Project': [
        'DIII-D', 'NSTX-U', 'NIF', 'SPARC', 'ITER Physics Support',
        'DIII-D', 'NSTX-U', 'NIF', 'SPARC', 'ITER Physics Support',
        'DIII-D', 'NSTX-U', 'NIF', 'SPARC', 'ITER Physics Support',
        'DIII-D', 'NSTX-U', 'NIF', 'SPARC', 'ITER Physics Support'
    ],
    'Facility': [
        'Perlmutter', 'Frontier', 'El Capitan', 'Aurora', 'Jupiter',
        'Perlmutter', 'Aurora', 'El Capitan', 'Aurora', 'LUMI',
        'Frontier', 'Aurora', 'El Capitan', 'Aurora', 'Henri',
        'Perlmutter', 'Frontier', 'El Capitan', 'Aurora', 'Jupiter'
    ],
    'Cores': [
        512, 1024, 2048, 4096, 8192,
        1024, 2048, 4096, 1024, 4096,
        2048, 4096, 8192, 4096, 2048,
        1024, 1024, 4096, 8192, 4096
    ],
    'RuntimeMinutes': [
        120, 180, 300, 450, 600,
        110, 150, 420, 300, 500,
        200, 240, 380, 400, 350,
        140, 160, 310, 490, 550
    ],
    'JobType': [
        'Simulation', 'Simulation', 'Post-processing', 'ML', 'Simulation',
        'Simulation', 'ML', 'Post-processing', 'Simulation', 'Simulation',
        'Simulation', 'Simulation', 'ML', 'ML', 'Post-processing',
        'ML', 'Simulation', 'Simulation', 'Post-processing', 'Simulation'
    ],
    'Energy_kWh': [
        150, 320, 600, 950, 1600,
        145, 500, 850, 720, 1500,
        300, 640, 1100, 980, 890,
        210, 350, 980, 1400, 1550
    ]
})

# Quick preview
df.head()
```

```python
# --- Heatmap: Average runtime by Facility and Project (with counts, ordered columns) ---

import numpy as np

# Desired column order to match the 2025 systems list
facility_order = [
    "Perlmutter", "Frontier", "El Capitan", "Aurora", "Jupiter",
    "LUMI", "Henri", "Leonardo", "Fugaku", "Shaheen III", "Setonix"
]

# Build pivot tables for mean and counts
mean_pivot = df.pivot_table(
    index="Project", columns="Facility", values="RuntimeMinutes", aggfunc="mean"
)
count_pivot = df.pivot_table(
    index="Project", columns="Facility", values="RuntimeMinutes", aggfunc="count"
)

# Reindex columns to the desired order; keep only those that appear
present_cols = [c for c in facility_order if c in mean_pivot.columns]
mean_pivot = mean_pivot.reindex(columns=present_cols)
count_pivot = count_pivot.reindex(columns=present_cols)

# Create a matching annotation array: "mean (n)" or blank for NaN
annot = mean_pivot.copy().astype(object)
for r in mean_pivot.index:
    for c in mean_pivot.columns:
        val = mean_pivot.loc[r, c]
        n = count_pivot.loc[r, c]
        annot.loc[r, c] = "" if (pd.isna(val) or pd.isna(n) or n == 0) else f"{val:.0f} ({int(n)})"

plt.figure(figsize=(12, 7))
sns.heatmap(
    mean_pivot, annot=annot, fmt="", cmap="YlGnBu",
    linewidths=0.5, linecolor="gray", cbar_kws={"label": "Average Runtime (minutes)"}
)
plt.title("Average Job Runtime by Facility and Fusion Project", fontsize=14, fontweight="bold")
plt.ylabel("Fusion Project", fontsize=12)
plt.xlabel("HPC Facility", fontsize=12)
plt.xticks(rotation=30, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# --- Optional: Row-normalized heatmap (each project scaled to 0‚Äì1 across facilities) ---
# This helps compare which facility is 'slow/fast' per project regardless of absolute minutes.

row_min = mean_pivot.min(axis=1)
row_max = mean_pivot.max(axis=1)
norm = (mean_pivot.sub(row_min, axis=0)).div((row_max - row_min).replace(0, np.nan), axis=0)

plt.figure(figsize=(12, 7))
sns.heatmap(
    norm, annot=False, cmap="viridis",
    linewidths=0.5, linecolor="gray", cbar_kws={"label": "Row-normalized Runtime (0‚Äì1)"}
)
plt.title("Relative Runtime by Facility (Row-normalized per Project)", fontsize=14, fontweight="bold")
plt.ylabel("Fusion Project", fontsize=12)
plt.xlabel("HPC Facility", fontsize=12)
plt.xticks(rotation=30, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```


## ü§ñ AI & Machine Learning in Fusion Research

Artificial Intelligence (AI) and Machine Learning (ML) are emerging as powerful tools in fusion energy R&D. These techniques are being used to:

- **Predict Disruptions**: ML models trained on DIII-D plasma shots predict disruptions with >90% accuracy (Kates-Harbeck et al., 2019).
- **Guide Magnetic Control**: Reinforcement learning (RL) agents are being tested to control plasma shape in real time.
- **Accelerate Simulation Workflows**: Surrogate models reduce compute time for gyrokinetic codes by orders of magnitude.

### üß™ SciDAC Initiatives
The DOE‚Äôs **Scientific Discovery through Advanced Computing (SciDAC)** program supports AI/ML infusion in MHD modeling, turbulence prediction, and materials degradation analysis.

## üèõÔ∏è Policy, Startups & Commercialization

Fusion is no longer a distant ambition‚Äîit's an active area of commercialization with strong federal backing.

### üîπ DOE Support
- **$1.4B in milestone-based grants** support U.S. fusion startups advancing toward pilot plants by the 2030s.
- **Clean energy tax credit eligibility** (post-IRA) supports demo plant financing.
- **TINEX** (Tokamak Innovation Network for Experimentation) is driving U.S. tokamak engineering innovation.

### üîπ Leading Private Companies
- **Commonwealth Fusion Systems (CFS)** ‚Äì MIT spinoff targeting net energy by 2026.
- **Helion Energy** ‚Äì Field-reversed configuration with private PPAs.
- **Zap Energy, TAE, General Fusion** ‚Äì Each pursuing alternate confinement approaches.

### üîπ Workforce Initiatives
- DOE and NSF fund workforce pipelines like the **FIRE Collaboratives** and **SCGSR Fellowship**.
- Broader inclusion goals target MSIs and underrepresented groups.
